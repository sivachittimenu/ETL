```python
df.columns=df.columns.str.replace("'","",regex=False).str.strip()
```

```python
df = df.replace(r"['\"]", "", regex=True) 
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df["NAV_DATE"]=pd.to_datetime(df["NAV_DATE"],errors="coerce")
```

```python
df.head()
```

```python
df["RECORD_ID"]=pd.to_numeric(df["RECORD_ID"],errors="coerce").fillna(0).astype("Int64")
df["RANK"]=pd.to_numeric(df["RANK"],errors="coerce").fillna(0).astype("Int64")
df["UNITS_HELD"]=pd.to_numeric(df["UNITS_HELD"],errors="coerce").fillna(0).astype("Int64")
df["NUM_INVESTORS"]=pd.to_numeric(df["NUM_INVESTORS"],errors="coerce").fillna(0).astype("Int64")
df["DAYS_SINCE_INCEPTION"]=pd.to_numeric(df["DAYS_SINCE_INCEPTION"],errors="coerce").fillna(0).astype("Int64")
```

```python
df.head()
```

```python
df["ISIN_NO"]=df["ISIN_NO"].fillna('0').astype("string")
df["PRODCODE"]=df["PRODCODE"].fillna('0').astype("string")
df["NAME"]=df["NAME"].fillna('0').astype("string")
df["SCHEME_TYPE"]=df["SCHEME_TYPE"].fillna('0').astype("string")
df.info()
df.head()
```

```python
df["SCHEME_TYPE"]=df["SCHEME_TYPE"].str.upper()
df.head()
```

```python
#converts into lower
df['PRODCODE_TRF']=df['PRODCODE'].str.lower() .str.strip()
df.head()
```

```python
df["NAV_DATE_TRF"] = pd.to_datetime(df["NAV_DATE"], format="%Y-%m-%d", errors="coerce")
df.head()
```

```python
# round 2 digits  
df['NAV_VALUE_TRF']=df['NAV_VALUE'].round(2)
df.head()
```

```python
#convert 
df['DIV_PERUNIT_TRF']=df['DIV_PERUNIT'].apply (lambda x: 1 if x==0 else ( 2 if x==1 else 3))
df.head()
```

```python
# convert digit
df['CORPDIVRATE_TRF']=df['CORPDIVRATE'].round(2)
df.head()
```

```python
mapping_rule = {
    "EQUITY(G)": "Equity",
    "EQUITY(S)": "Equity",
    "BOND": "Debt",
    "CASH": "MoneyMarket",
    "ULTRA LIQUID": "MoneyMarket",
    "FMP(CE)": "FMP",
    "INDEX FUND": "Index",
    "ELSS": "TaxSaving",
    "BALANCED": "Hybrid",
    "GILT": "Gilt",
    "GOLD FOF": "FOF",
    "FOF": "FOF",
    "STP": "ShortTermDebt"
}
df["SCHEME_TYPE_TRF"] = df["SCHEME_TYPE"].map(mapping_rule).fillna("Others")
df.head()
```

```python
import pandas as pd
import numpy as np

# Step 1: Trim spaces (leading/trailing)
df["ISIN_NO"] = df["ISIN_NO"].str.strip()

# Step 2: Convert to uppercase
df["ISIN_NO"] = df["ISIN_NO"].str.upper()

# Step 3: Replace empty strings with NaN
df["ISIN_NO"] = df["ISIN_NO"].replace("", np.nan)
df.head()
```

```python
df['RECORD_ID_TRF']=df.groupby("NAV_DATE").cumcount()+1
df.head()
```

```python
import pandas as pd
import numpy as np


conditions=[
    (df["UNITS_HELD"] == 0),
    (df["UNITS_HELD"] > 0) & (df["UNITS_HELD"] <= 1000),
    (df["UNITS_HELD"] > 1000) & (df["UNITS_HELD"] <= 100000),
    (df["UNITS_HELD"] > 100000)
]

labels = ["ZeroUnits", "LowUnits", "MidUnits", "HighUnits"]
df["UNITS_HELD_TRF"] = np.select(conditions, labels, default="Unknown")
df.head()
```

```python
df.tail()
```

```python
df["INVESTOR_DENSITY"] = df["NUM_INVESTORS"] / (df["UNITS_HELD"] / 1000)
df.head()
```

```python
df['DAYS_SINCE_INCEPTION_TRF']=df["DAYS_SINCE_INCEPTION"].apply(
    lambda x: "New" if x <= 365 
              else "0_5Y" if x <= 1825 
              else "5_10Y" if x <= 3650 
              else "10Y_PLUS")
df.head()
```

```python
df["RANK_TRF"] = df["RANK"].apply(
    lambda x: "Tier01" if 1 <= x <= 20
              else "Tier02" if 21 <= x <= 50
              else "Tier03" if 51 <= x <= 100
              else "Tier04" if 101 <= x <= 200
              else "Tier05")

df.head()
```
